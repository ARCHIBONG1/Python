STEP 5: Randomly selecting player


```python

def choose_first():
    
    import random
    choice_symbol = ""
    logic_control = True
    
    while logic_control == True:
        
        ques = input("Player 1, would you like to select your symbol or have the game select one for you. Yes or No?: ").upper() 
        if ques == "YES":
            while choice_symbol != "X" and choice_symbol != "O":
                choice_symbol = input("Please pick between X and O: ").upper()
                logic_control = False
        elif ques == "NO":
            if random.randint(0,1) == 0:
                choice_symbol = "X"
                logic_control = False
            else:
                choice_symbol= "O"
                logic_control = False
                
    return choice_symbol
```
