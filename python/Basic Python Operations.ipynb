### Pass, Break and Continue

These are keywords closely asociated with control flow.
The "Pass" keyword does nothing at all, the "Break" breaks out of the closest enclosing loop while the the "continue" keyword goes to the top of the closest enclosing loop. The examples below illustrate these

```python
odd_num = [1,3,5,7,9]
for x in odd_num:
    print("Hey")    # based on the syntax for the for loops, woithout providing a viable herein for the loop tpo execute, we woiuld encounter an error
```

```python
odd_num = [1,3,5,7,9]
for x in odd_num:
                        # No code, produces an error
print("Hey")
```

```python
odd_num = [1,3,5,7,9]
for x in odd_num:
    pass            # Pass keyword allows us to run the loop without providing a viable code an withour an error
print("Hey")
```

```python
name = "Archibong"
for letter in name:
    if letter == "i":
        continue         # Goes to the closest enclosing loop without printing i
    print(letter)
```

```python
name = "Archibong"
for letter in name:
    if letter == "i":
        break         # breaks out of the closest enclosing loop without printing i
    print(letter)
```

### List Comprehensions
They allow us to flatten out for loops

```python
num = [32,435,6,4,64]
l1 = [_ for _ in num]      # collapses the for loop into the list

print(l1)    # prints the new list
```

```python
num = range(1,30,2)    #generates a list of random numbers from 1 through 30 in steps of 2
l2 = [_**2 for _ in num]      # collapses the for loop

print(l2)    # prints the new list of squares
```

```python
num = range(1,30)    #generates a list of random numbers from 1 through 30, no step provided
l2 = [_**2 for _ in num if _ % 2 == 0]      # collapses the for loop into a single line of code

print(l2)    # prints the new list of squares
```

```python
celcius_temp = [34,1,45,100,7] # declares a list of temperatures in celsius
fahrenheit = [((9/5)*temp +32) for temp in celcius_temp]  #Allows us to collapse the loop calculating the temperature in fahrenheit
print(fahrenheit)  # prints the converted tenmperature
```

LESSER OF TWO EVENS: Write a function that returns the lesser of two given numbers if both numbers are even, but returns the greater if one or both numbers are odd


SOLUTION1

```python
def func1(num1,num2):
    
    if num1 % 2 == 0 and num2 % 2 == 0:   # both numbers are even
        if num1 > num2:
            return num2
        else:
            return num1
    elif num1 % 2 != 0 or num2 != 0:      # one or bot number are odd
        if num1 > num2:
            return num1
        else:
            return num2
```

```python
def func2(a,b):
    if a % 2 == 0 and b% 2 == 0:   # both numbers are even
        return min(a,b)
    else:                          # one or both numbers are odd
        return max(a,b)
```

OLD MACDONALD: Write a function that capitalizes the first and fourth letters of a name

```python
def func5(n):
    name = []                            # declares an empty list to hold our iterations
    for _ in range(len(n)):              # loops over the index of each of the letter in the string
        if _ == 0 or _ == 3:             # checks for the fist and fourth letter
            name.append(n[_].upper())    # Coverts thge first and fourth to upper case
        else:
            name.append(n[_].lower())    # holds all other letters as lower case
    return "".join(name)                 # joins the contents of the list name together
```

BLACKJACK: Given three integers between 1 and 11, if their sum is less than or equal to 21, return their sum. If their sum exceeds 21 and there's an eleven, reduce the total sum by 10. Finally, if the sum (even after adjustment) exceeds 21, return 'BUST'

```python
def func11(n1,n2,n3):
    
    total = n1+n2+n3
    if total <= 21:
        return total
    elif total > 21:
        if n1 == 11 or n2 == 11 or n3 == 11:
            if (total - 10) > 21:
                return "BUST"
            else:
                return (total - 10)
        else:
            return "BUST"
```

SUMMER OF '69: Return the sum of the numbers in the array, except ignore sections of numbers starting with a 6 and extending to the next 9 (every 6 will be followed by at least one 9). Return 0 for no numbers.

```python
def summer_69(args):
    total = 0
    add = True
    
    for _ in args:
        while add:       
            if _ != 6:
                total += _
                break
            else:
                add = False
        while not add:
            if _ != 9:
                break
            else:
                add = True
                break
                
    return total       
```

```python
summer_69([1, 3, 5])
```

```python
summer_69([4, 5, 6, 7, 8, 9])
```

```python
summer_69([2, 1, 6, 9, 11])
```
